if i have a value ->
N : 114 , P : 27 ,  K : 28 , temp : 24.99451759 , humidity : 57.93250202 , ph : 7.162802357 rainfall  : 192.8736822 will it be rice or coffee??

```python
# P(rice) ? 
# P(coffee) ?
lenrice = 0
lencoffee = 0  
for i in range(len(read)):
    if read['label'][i] == 'rice':
        lenrice+=1
    elif read['label'][i] == 'coffee':
        lencoffee+=1
print("no of rice :", lenrice," ","no of coffee  : ",lencoffee)

# probability of rice = probability of coffee = 100/100+100
PR =  lenrice/(lenrice+lencoffee)
PC =  lencoffee/(lenrice+lencoffee)

print("P(R) = ",PR)
print("P(C) = ",PC)
```

lets assume we have to plant rice, so
Score = P(R) x L(N|R) x L(P|R) x L(K|R) x  L(temp|R) x L(humidity|R) x L(ph|R) x L(rainfall|R)


```python
# Score for rice :

scoreRice = PR*LNR*LPR*LKR*LTR*LHR*LPhR*LRR
scoreCoffee = PC*LNC*LPC*LKC*LTC*LHC*LPhC*LRC
print(scoreRice,scoreCoffee)
```

```python
# if one of it is zero so taking log
from math import log

scoreRice = log(PR)+log(LNR)+log(LPR)+log(LKR)+log(LTR)+log(LHR)+log(LPhR)+log(LRR)
scoreCoffee = log(PC)+log(LNC)+log(LPC)+log(LKC)+log(LTC)+log(LHC)+log(LPhC)+log(LRC)
print(scoreRice,scoreCoffee)
```

```python
# since score of rice is greater that score of coffee, rice is the best crop to plant
```
